```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Amazon_stock_data.csv')

df.head()

df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Amazon Closing Price Over Time')
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Volume'])
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Amazon Trading Volume Over Time')
plt.show()

##
df['Year'] = df['Date'].dt.year

yearly_avg = df.groupby('Year')['Close'].mean().reset_index()

plt.figure(figsize=(12,6))
plt.bar(yearly_avg['Year'], yearly_avg['Close'])
plt.xlabel('Year')
plt.ylabel('Average Closing Price')
plt.title('Amazon Average Closing Price per Year')
plt.show()



```

I analyzed historical Amazon stock data using Python and Matplotlib. The first graph shows the long term trend in the closing price. The second graph highlights changes in trading volume over time. The third graph displays the average closing price per year, which allows for a clear comparison of Amazons yearly performance and shows how the stock has grown over time.
